## RAG

#### R --> Retrieval 
#### A --> Augmented 
#### G --> Generation

* First the document is stored in vwctor database . 
* After that we take prompt from the user .
* Revelant information is **Retrived** from the vector database on the base of the prompt
* The prompt and th context from the retrived documented is **Augemented** . 
* This augumenetde prompt is ued to **Generate** the respone from LLM.

```python
import os
import google.generativeai as genai 
```

```python
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
```

```python
from pypdf import PdfReader
```

```python
#Step1 : Configure the model 
# First lets configure the API Key
gemini_key = os.getenv('TestProject1')
genai.configure(api_key=gemini_key)
model = genai.GenerativeModel('gemini-2.5-flash-lite')

# Embdeding Model
embeding_model = HuggingFaceBgeEmbeddings(model_name='all-MiniLM-L6-v2')
```

```python
#pip install "numpy<2"
```

```python
#pip install faiss-cpu langchain-community sentence-transformers
```

```python
# Step 2: Lets read the pdf and extract the text from it

pdf_file = PdfReader(r"D:\GenAI_Applications\RAG_Based_ChatBox\sample.pdf")

raw_text = ''
for page in pdf_file.pages:
    text = page.extract_text()
    if text:
        raw_text = raw_text + text + '\n'
''
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter 
```

```python
# Step 3: Chunking 
# First we need to split the text 

splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap=200)  
chunks = splitter.split_text(raw_text)
```

```python
len(chunks)
```

```python
len(raw_text)
```

```python
import faiss 
from langchain_community.vectorstores import FAISS
```

```python
# Step4: Create the vector database (FAISS)
vector_store = FAISS.from_texts(chunks,embeding_model)
```

```python
#Step 5: Get the prompt from the user
prompt = 'Give the 5 line summary if the document.'
```

```python
#Step6: Retrieval (R)
retriever = vector_store.as_retriever(search_kargs = {'k':3})
retriever_docs = retriever.invoke(prompt)
```

```python
retriever_docs
```

```python
context = '\n'.join([d.page_content for d in retriever_docs])
```

```python
# Stpe 7 : Augmenting (A)
augmented_prompt = f'''
<Role> You are an helpful assistant uing RAG.
<Goal> Answer the question asked by the user . Here is the question: {prompt}
<Context> Here are th documents retrivered from the vetcor database to suport the answer which you have to generate {context}
'''
```

```python
#Step 8 : Generation (G)
respsone = model.generate_content(augmented_prompt)
print(respsone.text)
```
